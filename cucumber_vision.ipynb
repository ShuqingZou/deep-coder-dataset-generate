```python
import pickle
```

```python
data = None

with open("dataset/bickle100k.pickle", "rb") as file:
    data = pickle.load(file)
```

```python
print(data.metadata)
dataset = list(data.dataset)
print(f"len dataset={len(dataset)}")
print(f"dataset type={type(dataset)}")
print(f"dataset element type={type(dataset[0])}")
```

```python
from pprint import pprint

'''
@dataclasses.dataclass
class Entry:
    source_code: str
    examples: List[Example]
    attribute: Dict[str, bool]

@dataclasses.dataclass
class Example:
    """
    An I/O example

    Attributes
    ----------
    inputs : List[Primitive]
    output : Primitive
    """
    inputs: List[Primitive]
    output: Primitive

Primitive = Union[int, List[int]]
'''

print("SOURCE CODE")
print(dataset[0].source_code)
print()
print(f"num_examples={len(dataset[0].examples)}")
print()
print("first example")
pprint(dataset[0].examples[0], compact=True)
print()
```

```python
print("attributes  # which functions are present")
pprint(dataset[0].attribute, compact=True)
```
